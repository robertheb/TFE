```python
import pandas as pd
import pymongo
import numpy as np
from collections import Counter
import urllib
import urllib2
from bs4 import BeautifulSoup
from google import search
```

```python
comsumptionVoD = pd.read_csv('consumptionVoDCleanedForRecomendation.csv', delimiter=';')
videos = pd.read_csv('videosCleaned.csv', delimiter=',')
```

```python
comsumption = comsumptionVoD[['MOVIEID', 'UNIQUEUSERCODE']]
```

```python
comsumption
```

```python
comsumption['VIDEOPOINTS'] = 1
```

```python
#Agregamos antes de hacer el pivot
agregatedComsumption = comsumption.groupby(['UNIQUEUSERCODE', 'MOVIEID'], as_index=False).sum()
#CAMBIO FINAL, ELIMINAMOS VIDEOPOINTS Y AÑADIMOS 1
del agregatedComsumption['VIDEOPOINTS']
agregatedComsumption
```

```python
#CAMBIO PARA LA MATRIZ DE HA VISTO O NO HA VISTO LA PELI ---------------------------------------
agregatedComsumption['VIDEOPOINTS'] = 1
```

```python
agregatedComsumption
```

```python
#HACER PIVOT A comsumption Y SEGUIR DESDE AQUI CON LA MATRIZ DE CORRELACION HA VISTO LA PELICULA O NO LA HA VISTO

a = agregatedComsumption.pivot('UNIQUEUSERCODE', 'MOVIEID', 'VIDEOPOINTS')
```

```python
a
```

```python
comsumptionFilmPerUser = a.fillna(0)
movie_index = comsumptionFilmPerUser.columns
corr_matrix = np.corrcoef(comsumptionFilmPerUser.T)
corr_matrix.shape
```

```python
def functionRating(codVideo):
    favoured_movie_title = codVideo

    favoured_movie_index = list(movie_index).index(favoured_movie_title)

    P = corr_matrix[favoured_movie_index]

    #solo devolvemos las películas con la mayor correlación con Toy Story
    return list(movie_index[np.logical_and(P>0.2, P<1.0)])
```

```python
#Prueba de la función
functionRating(39472)
```

```python
userData = comsumption[comsumption.UNIQUEUSERCODE == 8899300200.0]

userVideos = userData['MOVIEID'].tolist()
```

```python
userVideos
```

```python
usario_recomendacion = Counter()
```

```python
for video in userVideos:
    videosRecomended = functionRating(video)
    for videoRec in videosRecomended:
        if videoRec not in userVideos:
            usario_recomendacion.update([videoRec])
```

```python
listaVideos = []
for x in range(0,10):
    videoForSearchTitle = usario_recomendacion.most_common(10)[x][0]
    videoData = videos[videos.CONTENTID == videoForSearchTitle]
    videoTitle = videoData['CONTENTTITLE'].item()
    listaVideos.append(videoTitle)
listaVideos
```

```python
def functionSearchTrailer(title):
    textToSearch = title + "trailer"
    query = urllib.quote(textToSearch)
    url = "https://www.youtube.com/results?search_query=" + query
    response = urllib2.urlopen(url)
    html = response.read()
    soup = BeautifulSoup(html)
    for vid in soup.findAll(attrs={'class':'yt-uix-tile-link'}):
        videoTrailers = 'https://www.youtube.com' + vid['href']
        return videoTrailers
```

```python
def functionSearchGoogleSynopsis(title):
    for url in search(title + "filmaffinity", num=1, start = 0, stop = 1):
        print(url)
```

```python
for title in listaVideos:
    print functionSearchTrailer(title)
```

```python
for title in listaVideos:
    functionSearchGoogleSynopsis(title)
```
